# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:black; overflow:hidden"><b>Import Libraries</b></div>

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings

# Filter warnings to keep the output clean
filterwarnings('ignore')

# Enable inline plotting for Jupyter
%matplotlib inline
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:black; overflow:hidden"><b>Load the Dataset</b></div>

```python
# Load the dataset
file_path = '/kaggle/input/country-comparison-dataset-usa-and-more/country_comparison_large_dataset.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:black; overflow:hidden"><b>Dataset Overview</b></div>

```python
# Get a summary of the dataset
df.info()

# Check for missing values
df.isnull().sum()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:black; overflow:hidden"><b>Handle Missing Data</b></div>

```python
# Separate numeric and non-numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns

# Fill missing values in numeric columns with the mean
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill missing values in non-numeric columns with the mode
df[non_numeric_cols] = df[non_numeric_cols].fillna(df[non_numeric_cols].mode().iloc[0])

# Verify there are no missing values
df.isnull().sum()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:black; overflow:hidden"><b>Basic Statistics</b></div>

```python
# Basic statistics for numeric columns
df.describe()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:black; overflow:hidden"><b>Correlation Matrix</b></div>

```python
# Filter the dataframe to include only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Generate the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Numeric Variables')
plt.show()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:black; overflow:hidden"><b>GDP and Life Expectancy Relationship</b></div>

```python
# Scatter plot to explore the relationship between GDP per Capita and Life Expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GDP per Capita (in USD)', y='Life Expectancy (Years)', hue='Country')
plt.title('GDP per Capita vs. Life Expectancy')
plt.show()
```

```python
# Scatter plot of Military Expenditure and CO2 Emissions
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Military Expenditure (in Billion USD)', y='CO2 Emissions (Million Metric Tons)', hue='Country')
plt.title('Military Expenditure vs CO2 Emissions')
plt.show()
```

```python
# Bar plot of Energy Consumption and Renewable Energy Share by Country
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Country', y='Energy Consumption (TWh)', hue='Renewable Energy Share (%)')
plt.title('Energy Consumption vs Renewable Energy Share')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter for a specific range of Poverty Rate (%) and select a random sample
filtered_df = df[df['Poverty Rate (%)'] <= 40]  # Filter poverty rate less than or equal to 40%
sampled_df = filtered_df.sample(n=50, random_state=1)  # Randomly select 50 records for a smaller dataset

# Box plot of HDI across different Poverty Rate ranges for the sampled data
plt.figure(figsize=(10, 6))
sns.boxplot(data=sampled_df, x='Poverty Rate (%)', y='Human Development Index (HDI)', hue='Country')
plt.title('HDI vs Poverty Rate (Sampled Data)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of Smartphone Adoption vs Internet Penetration
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Smartphone Adoption (%)', y='Internet Penetration (%)', hue='Country')
plt.title('Smartphone Adoption vs Internet Penetration')
plt.show()
```

```python
# Scatter plot of Urban Population (%) and Crime Rate (per 100,000)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Urban Population (%)', y='Crime Rate (per 100,000)', hue='Country')
plt.title('Urban Population vs Crime Rate')
plt.show()
```

```python
# Bar plot of Tourism Revenue and Number of International Visitors
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Country', y='Tourism Revenue (in Billion USD)', hue='Number of International Visitors (in Millions)')
plt.title('Tourism Revenue vs Number of International Visitors')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of Education Expenditure as % of GDP vs Literacy Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Education Expenditure as % of GDP', y='Literacy Rate (%)', hue='Country')
plt.title('Education Expenditure vs Literacy Rate')
plt.show()
```

```python
# Line plot of Road Network Length and Public Transport Usage over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Road Network Length (in km)', hue='Country')
plt.title('Road Network Length Over Time')
plt.show()

# Bar plot of Public Transport Usage
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='Public Transport Usage (%)')
plt.title('Public Transport Usage by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of Unemployment Rate (%) vs Inflation Rate (%)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Unemployment Rate (%)', y='Inflation Rate (%)', hue='Country')
plt.title('Unemployment Rate vs Inflation Rate')
plt.show()
```

```python
# Scatter plot of Poverty Rate (%) and Life Expectancy (Years)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Poverty Rate (%)', y='Life Expectancy (Years)', hue='Country')
plt.title('Poverty Rate vs Life Expectancy')
plt.show()
```

```python
# Bar plot of Military Expenditure vs Education Expenditure
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='Military Expenditure (in Billion USD)', hue='Education Expenditure as % of GDP')
plt.title('Military Expenditure vs Education Expenditure')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot of Crime Rate (per 100,000) across different countries
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Country', y='Crime Rate (per 100,000)')
plt.title('Crime Rate by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of Literacy Rate (%) vs Unemployment Rate (%)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Literacy Rate (%)', y='Unemployment Rate (%)', hue='Country')
plt.title('Literacy Rate vs Unemployment Rate')
plt.show()
```
